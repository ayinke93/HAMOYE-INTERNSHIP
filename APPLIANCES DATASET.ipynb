# Multiple linear regression

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sn
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
normalized_features= pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
```

```python
features=normalized_features.drop(['Appliances'], axis=1)
```

```python
target=normalized_features['Appliances']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     features, target, test_size=0.3, random_state=42)
```

```python
lm=LinearRegression()
```

```python
lm.fit(features,target)
```

```python
print(lm.intercept_,lm.coef_)
```

```python
lm.fit(X_train,y_train)
```

```python
y_test_a=lm.predict(X_test)
```

```python
y_train_a=lm.predict(X_train)
```

# Measuring the performance of the model

R-Squared Value

```python
from sklearn.metrics import r2_score
```

```python
R_squared= r2_score(y_test,y_test_a)
round(R_squared,3)
```

```python
r2_score(y_train,y_train_a)
```

Mean Absolute Error

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae=mean_absolute_error(y_test,y_test_a)
round(mae,3)
```

RESIDUAL SUM OF SQUARES

```python
import numpy as np
rss= np.sum(np.square(y_test-y_test_a))
round(rss,3)
```

ROOT MEAN SQUARE ERROR

```python
from sklearn.metrics import mean_squared_error
rmse=np.sqrt(mean_squared_error(y_test,y_test_a))
round(rmse,3)
```

MEAN SQUARED ERROR

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_test_a)
round(mse,3)
```

RIDGE REGRESSION

```python
lm_r = Ridge(alpha=0.4)
lm_r.fit(X_train,y_train)
```
