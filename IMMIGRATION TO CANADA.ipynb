```python

import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = pd.ExcelFile('Canada.xlsx')
df1= pd.read_excel(df,'Canada by Citizenship', skiprows=1)
df2=df1[['OdName',2009,2010,2011,2012,2013]]
df3=df2.drop(df2.index[196])
Country_data= df3.melt(id_vars=['OdName'],var_name='Year')
```

Migration is in large part related to the broader global economic, social, political and technological transformations that are affecting
a wide range of high-priority policy issues.As the processes of globalization deepen, these transformations increasingly shape our lives – in our workplaces, in our homes, in our social and spiritual lives – as we go about our daily routines. Increasing numbers of people are able to access information, goods and services from around the world because of the ongoing expansion in distance-shrinking technologies. 
In this project, I will be exploring the immigration into one of the most desirable immigration locations in the wrld, Canada. 

```python
choropleth= px.choropleth(Country_data, locations="OdName", locationmode= "country names",
              scope="world",
                    color="value",
                    title= "WORLD IMIGRATION TO CANADA",
                    hover_name="OdName",
                    color_continuous_scale=px.colors.sequential.Viridis, animation_frame='Year')
choropleth
```

The plot above shows a choropleth showing immigration into Canada from year 2009-2013 from variou Countries across the world. Something to notice is the high influx of immigrants form Asia to Canada. (data source- Kaggle.com)
